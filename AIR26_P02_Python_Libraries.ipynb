# 1. What is a Python Library?

A **library** is a collection of ready-made code.

Instead of writing everything from scratch, we can use libraries to perform common tasks.

```text
Without library: Write everything manually
With library: Use existing tested tools
```

## Simple analogy

A Python library is like a toolbox.

| Task | Library Example |
|---|---|
| Mathematical operations | `math`, `NumPy` |
| Plotting graphs | `Matplotlib` |
| Working with tables | `Pandas` |
| Scientific computing | `SciPy` |
| Machine Learning | `scikit-learn` |
| Computer Vision | `OpenCV` |

## Why libraries are useful in AI and Robotics

Libraries help us to:

- process sensor data,
- analyze experiment results,
- plot graphs,
- train machine learning models,
- process images and videos,
- write faster and cleaner code.

# 2. Library, Module, Package, Import, and Alias

These words are commonly used in Python.

| Term | Meaning |
|---|---|
| Library | Collection of useful code |
| Module | A single Python file containing functions/classes |
| Package | Collection of modules |
| Import | Bringing a library/module into our program |
| Alias | Short name used for a library |

To use a library, we write:

```python
import library_name
```

Sometimes we use short names called **aliases**:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

These aliases are standard conventions used by most Python programmers.

````python
# Importing a built-in library

import math

print(math.sqrt(25))
print(math.pi)
````

## Another import style

Sometimes we import only one function or value from a library.

Example:

```python
from math import pi
```

This allows us to use `pi` directly instead of writing `math.pi`.

````python
from math import pi

radius = 5
area = pi * radius ** 2

print("Area of circle:", area)
````

# 6. NumPy: Numerical Computing

`NumPy` is one of the most important libraries for AI and Machine Learning.

It is used for:

- arrays,
- mathematical operations,
- sensor data,
- matrices,
- numerical computation.

Import convention:

```python
import numpy as np
```

## Why NumPy is important for AI and Robotics

A robot may continuously receive sensor values such as:

```text
distance, speed, temperature, battery, position
```

NumPy helps us store and process such numerical data efficiently.

````python
import numpy as np

# A NumPy array stores numerical data efficiently.

sensor_readings = np.array([0.9, 0.7, 0.4, 1.2, 0.3])

print(sensor_readings)
print(type(sensor_readings))#nd means n-dimensions
````

````python
import numpy as np

sensor_readings = np.array([0.9, 0.7, 0.4, 1.2, 0.3])

print("Minimum distance:", np.min(sensor_readings))
print("Maximum distance:", np.max(sensor_readings))
print("Average distance:", np.mean(sensor_readings))
print("Standard deviation:", np.std(sensor_readings))
````

## Why NumPy is useful

With normal Python lists, many operations require loops.

With NumPy arrays, we can apply operations directly to all values.

This is called **vectorized operation**.

````python
import numpy as np

distances = np.array([0.9, 0.7, 0.4, 1.2, 0.3])

# Convert distance from meters to centimeters.
distances_cm = distances * 100

print("Distance in meters:", distances)
print("Distance in centimeters:", distances_cm)
````

## Indexing in NumPy

Indexing means accessing a value using its position.

Python indexing starts from `0`.

```text
Index:       0    1    2    3
Value:     10   20   30   40
```

````python
import numpy as np

battery_levels = np.array([100, 95, 90, 85, 80])

print("First battery reading:", battery_levels[0])
print("Last battery reading:", battery_levels[-1])
print("First three readings:", battery_levels[0:3])
````

````python
import numpy as np

# Robot positions: each row is [x, y]
positions = np.array([
    [0, 0],
    [1, 1],
    [2, 1],
    [3, 2]
])

print("Positions:")
print(positions)

print("Shape of array:", positions.shape)#.shape gives no of ros and no. of columns
````

# 8. Pandas: Working with Tables

`Pandas` is used to work with tabular data.

A table in Pandas is called a **DataFrame**.

Import convention:

```python
import pandas as pd
```

## Simple explanation

A Pandas DataFrame is like an **Excel table inside Python**.

```text
Rows    → observations / records
Columns → features / variables
```

In AI and Robotics, Pandas is useful for:

- datasets,
- experiment readings,
- CSV files,
- robot logs,
- machine learning data.

````python
import pandas as pd

# Creating a simple table using a dictionary.
# Each key becomes a column in the DataFrame.

data = {
    "time": [1, 2, 3, 4, 5],
    "distance": [0.9, 0.7, 0.4, 0.8, 1.1],
    "battery": [100, 96, 92, 88, 84]
}

df = pd.DataFrame(data)

df #for displaying the dataframe
````

## Understanding the DataFrame

In the table above:

- `time` is the time step,
- `distance` is the obstacle distance,
- `battery` is the robot battery level.

Each row represents one observation recorded by the robot.

````python
# Basic information about the DataFrame

print("First rows:")
print(df.head())

print("\nSummary statistics:")
print(df.describe())
````

````python
# Accessing one column

print(df["distance"])
````

## Adding a new column using decision logic

Now we will add a new column named `action`.

Rule:

```text
If distance < 0.5 → Stop
Otherwise        → Move
```

We will first use a normal function because it is easier for beginners.

````python
# Beginner-friendly method using normal function

def decide_action(distance):
    if distance < 0.5:
        return "Stop"
    else:
        return "Move"

df["action"] = df["distance"].apply(decide_action)

df
````

## Optional: Lambda Function

A `lambda` function is a short one-line function.

This is optional for beginners. You can skip this cell during the first reading.

The following cell does the same task using `lambda`.

````python
# Optional: Same logic using lambda function

df["action_lambda"] = df["distance"].apply(lambda d: "Stop" if d < 0.5 else "Move")

df
````

Robotics example:

# 14. Mini Lab: Robot Sensor Data Analysis

In this mini lab, we will combine:

- NumPy,
- Pandas,
- Matplotlib,
- decision-making logic.

Problem:

A robot has recorded obstacle distance and battery level over time.  
We want to:

1. store the data,
2. analyze the data,
3. decide robot action,
4. plot the results,
5. count how many times the robot moves or stops.

````python
# Step 1: Create robot sensor data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

time = np.arange(1, 11)
distance = np.array([1.2, 1.0, 0.8, 0.45, 0.4, 0.7, 1.1, 0.3, 0.6, 0.9])
battery = np.array([100, 97, 94, 90, 87, 84, 80, 76, 72, 68])

print("Time:", time)
print("Distance:", distance)
print("Battery:", battery)
````

````python
# Step 2: Store data in a DataFrame

robot_data = pd.DataFrame({
    "time": time,
    "distance": distance,
    "battery": battery
})

robot_data
````

````python
# Step 3: Add robot action

def robot_action(distance_value):
    if distance_value < 0.5:
        return "Stop"
    else:
        return "Move"

robot_data["action"] = robot_data["distance"].apply(robot_action)

robot_data
````

````python
# Step 4: Analyze the data

print("Average distance:", robot_data["distance"].mean())
print("Minimum distance:", robot_data["distance"].min())
print("Average battery:", robot_data["battery"].mean())
print("Number of stop decisions:", (robot_data["action"] == "Stop").sum())
````

````python
# Step 5: Plot obstacle distance over time

plt.plot(robot_data["time"], robot_data["distance"], marker="o")
plt.xlabel("Time Step")
plt.ylabel("Obstacle Distance (m)")
plt.title("Robot Obstacle Distance Over Time")
plt.grid(True)
plt.show()
````

````python
# Step 6: Plot battery level over time

plt.plot(robot_data["time"], robot_data["battery"], marker="o")
plt.xlabel("Time Step")
plt.ylabel("Battery Level (%)")
plt.title("Robot Battery Level Over Time")
plt.grid(True)
plt.show()
````

````python
# Step 7: Count how many times the robot moves or stops

action_counts = robot_data["action"].value_counts()

print(action_counts)

action_counts.plot(kind="bar")
plt.xlabel("Robot Action")
plt.ylabel("Count")
plt.title("Robot Action Count")
plt.show()
````
